```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
```

```python
def FF(period, theta = -5.65, wavelength = 1.55, n_trench = 1.44, n_tooth = 1.68):
    n_c = n_trench
    return (n_c * np.sin(np.radians(theta)) + (wavelength/period) - n_trench)/(n_tooth - n_trench)
```

```python
x_min = 870
x_max = 960
x = np.linspace(x_min, x_max, 100)
plt.grid(True, linestyle = ':')
plt.xlabel("period in $nm$")
plt.ylabel('filling factor')
plt.plot(x, FF(x))
plt.savefig('FF_vs_period.png')
```
