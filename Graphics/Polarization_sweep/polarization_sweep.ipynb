```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
angles = [-67.5, -45, -22.5, 0, 22.5, 45, 67.5]
transmission = [0.137, 0.174, 0.161, 0.106, 0.042, 0.006, 0.019]

ticklabelsize = 14
labelsize = 17

fig, ax = plt.subplots(1, 1, figsize=(6, 4)) 
ax.plot(angles, transmission, linewidth = 2)
ax.set_xlabel('polarization angle in degree', fontsize = labelsize)
ax.set_xticks(angles)
ax.set_xticklabels(angles)
ax.set_ylabel('transmission T', fontsize = labelsize)

ax.tick_params(axis = 'both',  labelsize = ticklabelsize)
ax.grid(linestyle = ':')
plt.show()
fig.savefig('polarization.pdf')
fig.savefig('polarization.jpg')
fig.savefig('polarization.png')
```
