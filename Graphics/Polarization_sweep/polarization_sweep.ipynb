```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
angles = [-67.5, -45, -22.5, 0, 22.5, 45, 67.5]
transmission = [0.137, 0.174, 0.161, 0.106, 0.042, 0.006, 0.019]
transmissionlabels = np.linspace(0.025, 0.175, 4)


ticklabelsize = 15
labelsize = 18

fig, ax = plt.subplots(1, 1, figsize=(5, 3.7)) 
ax.plot(angles, transmission, linewidth = 2.5)
ax.set_xlabel('Polarization angle in degree', fontsize = labelsize)
ax.set_xticks(angles)
ax.set_xticklabels(angles)

ax.set_ylabel('T', fontsize = labelsize)
ax.set_yticks(transmissionlabels)

ax.tick_params(axis = 'both',  labelsize = ticklabelsize)
ax.grid(linestyle = ':')
plt.tight_layout()
plt.show()
fig.savefig('polarization.pdf')
fig.savefig('polarization.jpg')
fig.savefig('polarization.png')
```
