```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
```

```python
def final2initial_FF(finalFF, FF):
    return np.add(FF, np.subtract(FF, finalFF))
```

```python
FF_880 =0.75
FF_890 = 0.67
FF_900 = 0.59
FF_910 = 0.52
```

```python
data_path880 = 'data\\up_T_880.txt'
data880 = pd.read_csv(data_path880, sep=",", header=None)
data880.columns = [0, 1]  # 0 is x and 1 is y
data880 = data880.drop([0])  # Drop the first row which contains the old header

# Convert the columns to numeric
data880[0] = pd.to_numeric(data880[0])
data880[1] = pd.to_numeric(data880[1])

data880[0] = final2initial_FF(data880[0], FF_880)


# figure2
data_path890 = 'data\\up_T_890.txt'
data890 = pd.read_csv(data_path890, sep=",", header=None)
data890.columns = [0, 1]  # 0 is x and 1 is y
data890 = data890.drop([0])  # Drop the first row which contains the old header

# Convert the columns to numeric
data890[0] = pd.to_numeric(data890[0])
data890[1] = pd.to_numeric(data890[1])

data890[0] = final2initial_FF(data890[0], FF_890)


# figure3
data_path900 = 'data\\up_T_900.txt'
data900 = pd.read_csv(data_path900, sep=",", header=None)
data900.columns = [0, 1]  # 0 is x and 1 is y
data900 = data900.drop([0])  # Drop the first row which contains the old header

# Convert the columns to numeric
data900[0] = pd.to_numeric(data900[0])
data900[1] = pd.to_numeric(data900[1])

data900[0] = final2initial_FF(data900[0], FF_900)

# figure 4
data_path910 = 'data\\up_T_910.txt'
data910 = pd.read_csv(data_path910, sep=",", header=None)
data910.columns = [0, 1]  # 0 is x and 1 is y
data910 = data910.drop([0])  # Drop the first row which contains the old header

# Convert the columns to numeric
data910[0] = pd.to_numeric(data910[0])
data910[1] = pd.to_numeric(data910[1])

data910[0] = final2initial_FF(data910[0], FF_910)
```

```python
tickslabelsize = 15
labelsize = 16
fontsize = 18
linewidth = 2

# Plot the data
fig, ax = plt.subplots(figsize=(8, 5), layout='constrained')
ax.plot(data880[0], data880[1], marker='', linestyle='-', label = r'$\Lambda = 880$', linewidth = linewidth)
ax.plot(data890[0], data890[1], marker='', linestyle='-', label = r'$\Lambda = 890$', linewidth = linewidth)
ax.plot(data900[0], data900[1], marker='', linestyle='-', label = r'$\Lambda = 900$', linewidth = linewidth)
ax.plot(data910[0], data910[1], marker='', linestyle='-', label = r'$\Lambda = 910$', linewidth = linewidth)


# Use scientific notation for x-axis tick labels if appropriate
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

ax.set_ylim(bottom= 0.375)
ax.set_xlim(left= 0.2)

# Increase label font size for readability and add a title
ax.tick_params(axis='both', labelrotation=0, labelsize=tickslabelsize)
ax.set_xlabel('initial filling factor', fontsize=fontsize)
ax.set_ylabel('transmission', fontsize=fontsize)
# ax.set_title('Transmission versus initial filling factor', fontsize=14)

ax.spines['bottom'].set_linewidth(1)  # Thicker line for x-axis
ax.spines['left'].set_linewidth(1)    # Thicker line for y-axis
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)

# # Rotate x-axis tick labels to prevent overlap
# ax.tick_params(axis='x', labelrotation=0, labelsize=10)

# Show grid
ax.grid(True, linestyle = ':')
ax.legend(fontsize = tickslabelsize)
plt.show()
fig.savefig('Up_T_4plots.png', dpi=300)
```
