```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_path = 'data\\smallResolution.txt'
data = pd.read_csv(data_path, sep=" ", header=None, skiprows= 29)
data = -1 * data.iloc[:, 1:]

x_ticks = pd.read_csv(data_path, sep=" ", header=None, skiprows= 3, nrows= 11)
x_ticks = x_ticks.iloc[:, 1:]
x_ticks = x_ticks.to_numpy()[:,0]*1000000
y_ticks = pd.read_csv(data_path, sep=" ", header=None, skiprows= 16, nrows= 11)
y_ticks = y_ticks.iloc[:, 1:]
y_ticks = y_ticks.to_numpy()[:,0]

# x_ticks = np.linspace(start=9.0e-05, stop=2.9e-04, num=data.shape[1])
# y_ticks = np.linspace(start=0.22, stop=0.547, num=data.shape[0])

# Calculate the positions for the ticks to be in the middle of the cells
# x_tick_positions = np.arange(0.5, data.shape[1], 1)
# y_tick_positions = np.round(np.arange(90 + 200/22, 290 + 200/22, 200/11), 3)

ymax = max(y_ticks)
ymin = min(y_ticks)
c = (ymax - ymin)/22
y_tick_positions = np.linspace(ymin + c, ymax -c, 11)

xmax = max(x_ticks)
xmin = min(x_ticks)
cx = (xmax - xmin)/22
x_tick_positions = np.linspace(xmin + cx, xmax -cx, 11)

Tmax = max(data)
print(Tmax)
print(range(1, 7))
all_values = []
for i in range(1, 12):
    all_values.extend(data[i].to_numpy())
print(len(all_values))
print(all_values)
print('max', min(all_values))
Tmin = min(all_values)
Tmax = max(all_values)
T_tick_positions = np.linspace(Tmin, Tmax, 8)


```

```python
extent = [min(x_ticks), max(x_ticks), min(y_ticks), max(y_ticks)]

fig, ax = plt.subplots(figsize=(8, 7), layout='constrained')
im = ax.imshow(data, extent=extent, aspect='auto') 

ax.set_xticks(x_tick_positions)
ax.set_yticks(y_tick_positions)

ax.set_xticklabels(x_ticks)
ax.set_yticklabels(y_ticks)

fig.colorbar(im, ax=ax, ticks= T_tick_positions)
plt.show()
```
