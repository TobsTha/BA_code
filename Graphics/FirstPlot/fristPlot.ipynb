```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def FF(period, wavelength = 1550, theta = -5.65, n_trench = 1.44, n_tooth = 1.68):
    n_c = n_trench
    return (n_c* np.sin(np.radians(theta)) + wavelength/period - n_trench)/(n_tooth - n_trench)

x_min = 860
x_max = 960
x = np.linspace(x_min, x_max, 100)
```

```python
fontsize = 20
fontsize_big = 25
plt.plot(x, FF(x), linewidth = 3)
plt.yticks(ticks = [0.1, 0.3, 0.5, 0.7, 0.9], labels = [0.1, 0.3, 0.5, 0.7, 0.9])
plt.tick_params(axis='both', which = 'major', labelsize = fontsize)
plt.xticks(ticks = [860, 910, 960], labels= [860, 910, 960])
plt.ylabel('FF', fontsize = fontsize_big)
plt.xlabel('$\\Lambda$ in $nm$', fontsize = fontsize_big)
plt.grid(True, linestyle = ':')
plt.tight_layout()
plt.savefig('FF_vs_period_afterSubmission.png')
```
