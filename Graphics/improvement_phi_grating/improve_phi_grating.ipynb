```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
phi = np.linspace(40, 50, 9)

data_path = data_path = 'improve_phi_grating.txt'
data_current = pd.read_csv(data_path, sep=",", header=None)
# print(data_current)
transmission = pd.to_numeric(data_current.iloc[1:,1].T)*(-1)
print(transmission)
print(phi)
```

```python
ticklabelsize = 16
labelsize = 19

fig, ax = plt.subplots(1, 1, figsize=(6, 4)) 
ax.plot(phi, transmission, linewidth = 2)
ax.set_xlabel('psi', fontsize = labelsize)
ax.set_xticks(phi)
ax.set_xticklabels(phi.astype(int))
ax.set_ylabel('transmission T', fontsize = labelsize)

ax.tick_params(axis = 'both',  labelsize = ticklabelsize)
ax.grid(linestyle = ':')
# plt.tight_layout()
plt.show()
fig.savefig('psi_transmission.png', dpi=300, bbox_inches='tight')
```
