```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def ellipse_like(phi, wavelength = 0.00000155, n_eff = 1.56, n_c = 1.44, theta = -5.65, number_of_grating = 50):
    r = wavelength/(n_eff - np.cos(np.radians(phi))*np.sin(np.radians(theta))* n_c) * number_of_grating
    return r
```

```python
phi_grating = 90

number_angles = 100
angles = np.linspace(-phi_grating, phi_grating, number_angles)
radii = ellipse_like(angles)
radii0 = ellipse_like(angles, theta= 0)
radii8 = ellipse_like(angles, theta=-8)

# Create a polar plot
plt.subplot(111, polar=True)

# Plot data
# plt.plot(np.radians(angles), radii0, label = '$\\theta = -0^\\circ$')
plt.plot(np.radians(angles), radii, label = '$\\theta = -5.65^\\circ$, perfect shape')
# plt.plot(np.radians(angles), radii8, label = '$\\theta = -8^\\circ$')

# plot circle:
# R8 = ellipse_like(0, theta = -8)
# radiiR = np.ones(number_angles)*R8
# plt.plot(np.radians(angles), radiiR, label = '$\\theta = -8^\\circ$, circular')

R565 = ellipse_like(0)
radiiR = np.ones(number_angles)*R565
plt.plot(np.radians(angles), radiiR, label = '$\\theta = -5.65^\\circ$, circular')

# plt.ylabel('radius in $\\mu m$')

plt.grid(linestyle = ':')
plt.legend()
plt.show()
plt.savefig('polarPlot.png')
phi_difference = 35
print('8 degree difference at phi = {} degree:'.format(phi_difference), ellipse_like(phi_difference, theta=-8) - R8)
print('5.65 degree difference at phi = {} degree:'.format(phi_difference), ellipse_like(phi_difference) - R565)
```
