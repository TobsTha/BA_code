```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_path = 'data\\up_T.txt'
data_current = pd.read_csv(data_path, sep=",", header=None)
# print(data_current)
transmission = pd.to_numeric(data_current.iloc[1:,1].T)
period = np.linspace(880, 900, 11)
period_square = period**2
print(transmission)
print(period_square)
```

```python
ticklabelsize = 14
labelsize = 17

fig, ax = plt.subplots(1, 1, figsize=(6, 4)) 
ax.plot(period, transmission, linewidth = 2)
ax.set_xlabel('period', fontsize = labelsize)
ax.set_xticks(period)
ax.set_xticklabels(period.astype(int))
ax.set_ylabel('transmission T', fontsize = labelsize)

ax.tick_params(axis = 'both',  labelsize = ticklabelsize)
ax.grid(linestyle = ':')
# plt.tight_layout()
plt.show()
fig.savefig('up_T.png', dpi=300, bbox_inches='tight')
```
